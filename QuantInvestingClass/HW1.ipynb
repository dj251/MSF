```python
daytime_return = (spy['Close'] - spy['Open']) / spy['Open']
nighttime_return = (spy['Open'] - spy['Close'].shift(1)) / spy['Close'].shift(1)
print("day time returns:",daytime_return.mean())
print("night time returns:",nighttime_return.mean())
```

```python
fullday_returns = (spy['Close'] - spy['Close'].shift(1)) / spy['Close'].shift(1)
print(fullday_returns.mean())
```

```python
fullday_returns_adj = (spy['Adj Close'] - spy['Adj Close'].shift(1)) / spy['Adj Close'].shift(1)
print(fullday_returns_adj.mean())
```
